## Exploratory Data Analysis

Import Libraries

```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import ast
import matplotlib.ticker
import seaborn as sns
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
```

Job Count for Data Analyst in US Based on Job Location  

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country'] == "United States" )].copy()
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame() #to convert into data frame
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r',legend=False)
for i, txt in enumerate(df_plot['count']):
    plt.text(txt - 500, i, str(txt), va='center', color = 'white')
plt.title("Job Location Based on Job Count for Data Analyst in US")
plt.xlabel("Job Count")
plt.ylabel("Job Location")
#plt.figure(figsize=(30,30))
plt.show()
```

Pie Chart showing Ratio of WFH Status, Degree Requirement, Health Insurance Offered in job Posting


```python
dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance' : 'Health Insurance Offered'
}

fig,ax = plt.subplots(1,3)

for i,(column,title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(),startangle = 90,labels= [False,True],autopct = "%1.1f%%" )
    ax[i].set_title(title)
plt.tight_layout()
plt.show()
```

Number of Job Posted per Company for Data Analyst in US

```python
df_DA_US_company = df_DA_US['company_name'].value_counts().head(10).to_frame()
df_DA_US_company
```

```python
sns.barplot(data=df_DA_US_company, x='count', y='company_name',hue='count',palette='dark:b_r',legend=False)
for i,txt in enumerate(df_DA_US_company['count']):
    plt.text(txt - 70 ,i,str(txt),va='center',color='white')
plt.title("Number of Job Posted per Company for Data Analyst in US")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```
